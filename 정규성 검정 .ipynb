```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
A=np.array([1,2,3,2,1,2,3,1,2,3,1,2,3,3]) # 랜덤하게 놓음
B=np.random.standard_normal(1000) # 평균이 0이고 표준편차가 1인 정규 분포 (가우시안 분포) 에서 난수를 생성합니다.
```

```python
plt.hist(A)
```

```python
plt.hist(B)
```

```python
normal_test_A, p_A=stats.shapiro(A) # 정규성 검정
normal_test_B, p_B=stats.shapiro(B)# 정규성 검정
print("A의 p 값:",p_A,"B의 p 값:",p_B) # p의값이 0.05보다 작으면 정규성 보장 x
# p의값이 0.05보다 크면 정규성 보장 o
```
