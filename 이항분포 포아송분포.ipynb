```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial as fact
import math
```

```python
#이항 분포
def bin_prob(n,k,p):
  nck=fact(n)/(fact(k)*fact(n-k))
  expect=nck*math.pow(p,k)*math.pow((1-p),(n-k))
  return expect
```

```python
expect_=[]
n=100
p=0.52
for i in range(n):
  expect_.append(bin_prob(n,i,p))

plt.bar(np.arange(n),expect_)
#plt.bar((np.arange(n)-n*p)/math.sqrt(n*p*(1-p)),expect_)
```
