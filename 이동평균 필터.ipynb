```python
import matplotlib.pyplot as plt
```

```python
A=[1074,1079,1083,1082,1077,1085,1082,1079,1073,1083,1089,1093,1094,1100,1111]
```

```python
def MAF(x):
    n=len(x)
    y=[]
    for i in range(n-2):
        temp=(x[i]+x[i+1]+x[i+2])/3
        y.append(temp)
    return y
fittered_A=MAF(A)
```

```python
X=range(len(A))
X1=range(len(fittered_A))
```

```python
plt.plot(X,A)
plt.plot(X1,fittered_A,'r') 
```
